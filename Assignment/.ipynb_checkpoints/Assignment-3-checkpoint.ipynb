# Data Cleaning Task of Income Classification

## 1. Load the data

```python
#we have used census dataset for income classification
#import all the libraries used in our dataset
import pandas as pd
import numpy as np
```

```python
#load csv file into notebook
df=pd.read_csv("../Dataset/adult.csv")
#first five rows from csv
df.head()
```

```python
#last five rows from dataset
df.tail()
```

## 2. Describe the data

```python
# it is used to describe statistical analysis of data
df.describe()
```

```python
# this gives the information about every column of our dataset
df.info()
```

```python
# check whether the our dataset have null values or not
df.isnull().sum()
```

```python
# count unique values from each column
df.nunique(axis=0)
```

```python
# return all the columns in dataset
Column=list(df.columns)
```

```python
Column
```

```python
#shown each columns and their unique values in dataset
for x in Column:
    print(x,df[x].unique())
    print()
```

## 3. Remove Insignificant Column

```python
# drop Fnlwgt, Education, Relationship,Race ,Gender column from our data set
df.drop(df.columns[[2,3,5,7,8]],axis=1,inplace=True)
df.head()
```

```python
#store all columns in our dataset
Column1=list(df.columns)
```

```python
Column1
```

## 4. Check Missing value for each columns

```python
#check null value from each column and sum of all true values
df.isnull().sum() 
```

```python
# Replace each ? from our dataset as missing values
df.replace(to_replace='?',value=np.nan,inplace=True)
```

```python
# after replacing ? to null values check null values in our dataset
df.isnull().sum()
```

```python
df.head()
```

```python
# fill all none values to nan and drop all the rows having none values
df.fillna(np.nan)
clean_df=df.dropna()
clean_df.head()
```

```python
# deleted rows from dataset after removing none values
print(len(df)-len(clean_df))
```

```python
# observe each column and their unique values and found no ? and none values in our dataset
for x in Column1:
    print(x,df[x].unique())
    print()
```

## 5. Formatting your Data â€“ making data types compaitable with other data types

```python
clean_df.head()
```

```python
clean_df.info()
```

```python
# make a copy of our clean data
modified_df=clean_df.copy()
```

```python
modified_df.head()
```

```python
#modified all the values of income having "<=50K" to 0
modified_df.loc[modified_df['income']=='<=50K','income']=0
```

```python
modified_df.head()
```

```python
#modified all the values of income having ">50K" to 1
modified_df.loc[modified_df['income']=='>50K','income']=1
modified_df.head()
```

```python
# rename educational-num to education-num
modified_df['education-num']=modified_df['educational-num']
```

```python
modified_df.head()
```

```python
# deleted educational-num column
del modified_df['educational-num']
```

```python
modified_df.head()
```

```python
# create a copy of modified data
formatted_df=modified_df.copy()
```

```python
formatted_col=formatted_df.columns
```

```python
print(formatted_col,len(formatted_col))
```

```python
formatted_df.head()
```

```python
formatted_df.info()
```

```python
# save all the formatted data into new csv file
formatted_df.to_csv('adult_data_formatted.csv')
```

## Graphs

```python
# import matplotlib and seaborn into notebook to make graphs
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Size=formatted_df['income'].size
Size
```

```python
# this shows all the rows have income greater than 50K
Greater_50K=formatted_df['income'].sum()
Greater_50K
```

```python
# this shows all the rows have income less than 50K
Less_50K=Size-Greater_50K
Less_50K
```

```python
# plot pie graph from income and peoples
distribution = np.array([Less_50K, Greater_50K])
labels = ['Below 50K', 'Above 50K']
explode = [0, 0.1]
plt.pie(x = distribution, explode=explode, labels=labels, autopct='%1.2f%%', shadow=True, startangle=45)
plt.show()
```

```python
Edu_data=formatted_df['education-num']
Edu_data.head()
```

```python
bins = np.arange(Edu_data.max()+1)
bins
```

```python
# plot histogram btw education num and people
plt.hist(x = Edu_data, bins = bins, color = 'green')
plt.xlabel('Education-Number')
plt.ylabel('People')
plt.show()
```

```python
sns.distplot(Edu_data, hist = False, color = 'm')
```
